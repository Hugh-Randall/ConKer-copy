```python
import numpy as np
from astropy.io import fits

s_ra_array = []
s_dec_array = []
s_z_array = []
s_wts_array = []

n = 0

while n < 18:
    SGC = fits.open('/global/cfs/cdirs/desi/survey/catalogs/Y1/mocks/SecondGenMocks/EZmock/FFA/mock1/ELG_LOP_ffa_SGC_{}_clustering.ran.fits'.format(n))[1].data

    weight_s = np.array(SGC['WEIGHT'])*np.array(SGC['WEIGHT_FKP'])
    s_ra_array.extend(SGC['RA'])
    s_dec_array.extend(SGC['DEC'])
    s_z_array.extend(SGC['Z'])
    s_wts_array.extend(weight_s)
    n+=1

out_cols = []
out_cols.append(fits.Column(name='ra',array=s_ra_array,format='E'))
out_cols.append(fits.Column(name='dec',array=s_dec_array,format='E'))
out_cols.append(fits.Column(name='z',array=s_z_array,format='E'))
out_cols.append(fits.Column(name='wts',array = s_wts_array,format='E'))
out_table = fits.BinTableHDU.from_columns(out_cols)

fname_s = '/global/homes/b/blevi2/EZmock_ELG/ELG_LOP_ffa_SGC_m1_clustering.ran.fits'

out_table.writeto(fname_s,overwrite=False)
```
